```python
x = np.array([1, 2, 3, 4])
y = x * 2
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Random Chart")
plt.plot(x, y)
plt.show()
```

```python
#Changing the style of line chart
x = np.array([1, 2, 3, 4])
y = x * 2
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Random Chart")
plt.plot(x, y, '-.')
plt.show()
```

```python
#Multiple charts
#Changing the style of line chart
x = np.array([1, 2, 3, 4])
y = x * 2
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Random Chart1")
plt.plot(x, y, '-.')
plt.show()

plt.figure() #this creates a space for another chart
x1 = np.array([1, 2, 3, 4])
y1 = x1 *3
plt.plot(x1, y1)
plt.show()
```

```python
#multiple charts on same axis
#Multiple charts
#Changing the style of line chart
x = np.array([1, 2, 3, 4])
y = x * 2
plt.plot(x, y, '-.')

x1 = np.array([1, 2, 3, 4])
y1 = x1 *3
plt.plot(x1, y1)

plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Random Chart1")
plt.show()
```

```python
#filling the area between two plots
x = np.array([1, 2, 3, 4])
y = x * 2
plt.plot(x, y, '-.')

x1 = np.array([1, 2, 3, 4])
y1 = x1 *3
plt.plot(x1, y1)

plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Random Chart1")

plt.fill_between(x, y, y1, color = 'yellow', alpha = 0.5)
plt.show()
```

```python
#Heatmap
df =pd.read_csv("GS.csv")
df = df.set_index("Date")
```

```python
df["Date"] = df.index
#converting to datetime type
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Month"] = df["Date"].dt.month
df["Year"] = df["Date"].dt.year
```

```python
#creating a pivot table with "Open" col
table_df = pd.pivot(df, values=["Open"], index = ["Month"], columns = ["Year"], fill_value = 0)
```
