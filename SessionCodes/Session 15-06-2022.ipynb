Correlation Matrix Plot

```python
names = ['Pregnancies', 'Glucose', 'Blood Presure', 'Skin Thickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Class']
df_corr_plots = pd.read_csv(r"pima_indians_diabetes.csv", names = names)
```

```python
import numpy as np
```

```python
correlation = df_corr_plots.corr()
fig = plt.figure(figsize = (25, 15))
ax = fig.add_subplot(1, 1, 1)
cax = ax.matshow(correlation, vmin = 1, vmax = 1)
fig.colorbar(cax)
ticks = np.arange(0, 9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
#pearson correlation
correlation = df_corr_plots.corr(method = 'pearson')
fig = plt.figure(figsize = (25, 15))
ax = fig.add_subplot(1, 1, 1)
cax = ax.matshow(correlation, vmin = 1, vmax = 1)
fig.colorbar(cax)
ticks = np.arange(0, 9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
