```python
import pandas as pd
from pandas import Series
```

```python
arr = Series([1, 4, 5, 6 ,78])
print(arr)
```

Changing the index

```python
arr = Series(["Mango", "Pineapple", "Apple", "Banana", "Kiwi"], index = ["M", "P", "A", "B", "K"])
print(arr)
```

```python
print(f"The values in the series are: {arr.values}")
print(f"The indices in the series are: {arr.index}")
```

```python
arr = Series(["Mango", "Pineapple", "Apple", "Banana", "Kiwi"], index = ["M", "P", "A", "B", "K"])
print(arr)
```

Group By

```python
#get total sales by each shipping mode
df1.groupby('Ship Mode').sum()
```

```python
#get average sales by each shipping mode
df1.groupby('Ship Mode').mean()
```

How to group by on multiple columns

```python
df1.groupby(['Ship Mode', 'Region']).sum()
```

How to get limited columns from the above group by

```python
df2 = df1.groupby(['Ship Mode', 'Region']).sum().Sales
```
